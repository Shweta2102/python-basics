# Q9) What is type casting in Python?

   
  type casting in python refers to changing the type. it is of two types: implicit and explicit. explicit means during an operation itself we change its type without defining into another variable but its original type is same
  a=5.5
  b=10
  c=int(a)+b  #explicitly converts float to int.

  *here the type of a is float but it is used as an integer just for the operation in c.*

  implicit means when we change its type and store it in another variable.
  key point: integer can be taken as a string but a string cannot be taken as an integer.

  for e.g.
  a=input()

  *given input 20. input takes be default values as a string but we want to convert it into an integer then we will do typecasting.*

  b=int(a) #implicitly converts srint to integer
  type(b)

  *now the type will be integer*
